### Ratio of images in the dataset.

```python
def get_an_img_and_calculate_ratio(img_file_name):
    img = cv2.imread(img_file_name, cv2.IMREAD_COLOR)
    height, width = img.shape[:2]
    ratio = height / width
    
    return ratio, [height, width]

def get_img_list(base_dir):
    """
    excluding matte images
    """
    return [os.path.join(base_dir, s) for s in os.listdir(base_dir) if s.split('.')[0][-5:] != "matte"]    
    
    
```

```python
train_base = "../dataset/selfie/training/"
test_base  = "../dataset/selfie/testing/"

train_ratio_list = list(map(get_an_img_and_calculate_ratio, get_img_list(train_base)))
test_ratio_list  = list(map(get_an_img_and_calculate_ratio, get_img_list(test_base)))
```
